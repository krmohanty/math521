```matlab
A=imread('CTimage.jpg');
```

```matlab
imshow(A);
```

```matlab
A = rgb2gray(A);
```

```matlab
avg3 = fspecial('average');   
```

```matlab
A1 = imfilter(A,avg3);
```

```matlab
imshow(A1);
```

```matlab
A1=double(A1);
```

```matlab
lapFilter = [0 -1 0; -1 4 -1; 0 -1 0];
```

```matlab
A2 = conv2(A, lapFilter,'same');
```

```matlab
A2=double(A2);
```

```matlab
A3 = A1+A2;
```

```matlab
imshow(uint8(A2));
```

```matlab
imshow(uint8(A3));
```
