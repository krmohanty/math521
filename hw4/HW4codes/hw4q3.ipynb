# HW4 Computing Problem 3

```matlab
x = (0:1/250:1-1/250)';
```

```matlab
size(x)
```

```matlab
for i = 1:10
    coef = (rand(10,1)-0.5)*2;
    Xe(:,i) = sin(x*2*pi*(1:10))*coef;
end
```

```matlab
size(Xe)
```

#### The original signal

```matlab
for i = 1:10
    subplot(5,2,i);plot(Xe(:,i));
end
```

#### Create correlated noise of rank 3 and add to signal

```matlab
X = Xe+(rand(length(x),3)-0.5)*2*rand(3,10);
```

```matlab
size(X)
```

```matlab
for i = 1:10
    subplot(5,2,i);plot(X(:,i));
end
```

#### Estimate the noise by taking the derivative

```matlab
N = X(2:end,:)-X(1:end-1,:);
```

```matlab
size(N)
```

#### We solve the generalized eigenvalue problem 

```matlab
[psi, e] = eig(N'*N, X'*X);
```

```matlab
plot(diag(e),'o')
xlabel('')
ylabel('\lambda')
```

```matlab
phi = X*psi;
```

```matlab
Y = phi(:,1:7)*(phi(:,1:7)'*X);
```

```matlab
x0=10;
y0=10;
width=550;
height=800
set(gcf,'position',[x0,y0,width,height])
for i = 1:10
subplot(5,2,i);plot(x,Y(:,i),'r');hold on;plot(x,Xe(:,i));hold on;plot(x,X(:,i))
end
```

```matlab
Y = phi(:,1:1)*(phi(:,1:1)'*X);
figure() ;
for i = 1:10
subplot(5,2,i);plot(x,Y(:,i));hold on;plot(x,Xe(:,i));hold on;plot(x,X(:,i));
legend('filtered data','original data','noisy data');
end
sgtitle(strcat("D= ",num2str(1)));
```

```matlab
%%
Y = phi(:,1:2)*(phi(:,1:2)'*X);
figure() ;
for i = 1:10
subplot(5,2,i);plot(x,Y(:,i));hold on;plot(x,Xe(:,i));hold on;plot(x,X(:,i));
legend('filtered data','original data','noisy data');
end
sgtitle(strcat("D= ",num2str(2)));
```
